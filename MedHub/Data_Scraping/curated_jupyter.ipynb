```python
import os
from os import listdir
from os.path import isfile, join
from disease_match_query import *
```

```python
def filelist(root, skip_files ='resources', dirname = False):
    """
    Get the list of filenames from the medline directory. Ignore filenames that contain 'Resources'
    """
    result =[]
    for path, subdirs, files in os.walk(root):
        for name in files:
            if not name.startswith('.') and skip_files not in name:
                result.append(os.path.join(path, name))
                
    if dirname == False:         
        result = [filepath.split("/")[-1] for filepath in result]
    
    return result
```
