```python
from os import listdir
from os.path import isfile, join
import gensim
from sklearn.feature_extraction.text import *
import sys, os, numpy, re, string
import nltk
from sklearn.feature_extraction.stop_words import *
from collections import Counter
```

```python
root_dir = "/Users/shivakanththudi/Desktop/appdev/data/"
specific_file = "/Users/shivakanththudi/Desktop/appdev/data/Scabies.txt"
```

```python
def tokenize_custom(text):
    """
    Tokenize text and return a non-unique list of tokenized words
    found in the text. Normalize to lowercase, strip punctuation,
    remove stop words, drop words of length < 3.
    """
    text = text.lower()
    subs = re.compile('[' + re.escape(string.punctuation) + '0-9\\r\\t\\n' + ']')
    processed_text = subs.sub(' ', text)
    words = nltk.word_tokenize(processed_text)
    words = [word for word in words if len(word) >=3 and word not in ENGLISH_STOP_WORDS]
    return words

def stemwords(words):
    """
    Given a list of tokens/words, return a new list with each word
    stemmed using a PorterStemmer.
    """
    stemmer = nltk.stem.porter.PorterStemmer()
    stemmed_list = [stemmer.stem_word(word) for word in words]
    return stemmed_list
```

```python
def gettext(text):

    """
    Parse text and remove JavaScript and references
    """
    return text.replace("To use the sharing features on this page, please enable JavaScript.", '').split("References")[0]
```

```python
def tokenizer_custom(text):
    return stemwords(tokenize_custom(text))
```

```python
def filelist(root, skip_files ='resources'):
    result =[]
    for path, subdirs, files in os.walk(root):
        for name in files:
            if not name.startswith('.') and skip_files not in name:
                result.append(os.path.join(path, name))
    return result
```

```python
tfidf = TfidfVectorizer(decode_error = 'ignore',input='filename', # argument to transform() is list of files
                        analyzer='word',
                        tokenizer=tokenizer_custom,  # tokenize, stem
                        preprocessor = gettext,
                        stop_words='english') # strip out stop words
```

```python
filenames_list = filelist(root_dir)

sparse_matrix = tfidf.fit_transform(filenames_list)
vocab = tfidf.get_feature_names()
```

```python
def top_terms(disease, threshold, term_doc_matrix, vocabulary):

    specific_file = "/Users/shivakanththudi/Desktop/appdev/data/%s.txt" % disease 
    specific_file_index = filenames_list.index(specific_file)

    dtm = numpy.nonzero(term_doc_matrix)
    file_index = dtm[0]
    tuples = [(dtm[0][i],dtm[1][i]) for i in range(len(file_index)) if file_index[i] == specific_file_index]

    vocab = numpy.array(vocabulary)

    sorted_list =  sorted([(vocab[each[1]], sparse_matrix[each[0],each[1]],3) for each in tuples if sparse_matrix[each[0],each[1]] >= threshold],
    key = lambda x: -x[1])

    for each in sorted_list:
        print each[0], '%5.3f' %(each[1])
```

```python
top_terms("Scabies", 0.09, sparse_matrix, vocab)
```

### DOC2VEC stuff

```python
LabeledSentence = gensim.models.doc2vec.LabeledSentence

docLabels = []
docLabels = [f for f in listdir("/Users/shivakanththudi/Desktop/appdev/data") if f.endswith('.txt')]
```

```python
data = []
for doc in docLabels:
    if 'resources' not in doc:
        f = open("/Users/shivakanththudi/Desktop/appdev/data/" + doc, 'r')
        data.append(f.read())
        f.close()
```

```python
# data = data[:10]
# docLabels = docLabels[:10]
```

```python
class LabeledLineSentence(object):
    def __init__(self, doc_list, labels_list):
       self.labels_list = labels_list
       self.doc_list = doc_list
    def __iter__(self):
        for idx, doc in enumerate(self.doc_list):
            yield LabeledSentence(words=doc.split(),tags=[self.labels_list[idx]])
```

```python
it = LabeledLineSentence(data, docLabels)
```

```python
model = gensim.models.Doc2Vec(workers=8) # use fixed learning rate
model.build_vocab(it)
for epoch in range(10):
    model.train(it)
    model.alpha -= 0.002 # decrease the learning rate
    model.min_alpha = model.alpha # fix the learning rate, no deca
    model.train(it)
```

```python
model.save("doc2vec.model")
```

```python
model.most_similar("white tongue")
```

```python
tokens = "white tongue".split()

new_vector = model.infer_vector(tokens)
sims = model.docvecs.most_similar([new_vector])
```
