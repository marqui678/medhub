```python
import pandas as pd, urllib, urllib2, json, numpy as np
pd.options.mode.chained_assignment = None
import random
from time import sleep

```

```python
# Simulate Ratings
ratings = [3.0, 3.5, 4.0, 4.5, 5.0]
sf_npi_data['Rating'] = np.random.choice(ratings, sf_npi_data.shape[0])

# Map Internal Medicine and Family Practice to General Practice
sf_npi_data = sf_npi_data.replace('INTERNAL MEDICINE', 'GENERAL PRACTICE', regex = True)
sf_npi_data = sf_npi_data.replace('FAMILY PRACTICE', 'GENERAL PRACTICE', regex = True)
```

```python
def address_to_coordinates(address):
    """
    Given a text address, returns the corresponding latitude and longitude

    Input: Address (String)
    
    Output: Latitude, Longitude
    """    
    params = {
            'address' : address,
            'sensor' : 'false',
    }  
    url = 'https://maps.google.com/maps/api/geocode/json?' + urllib.urlencode(params) + '&key=AIzaSyDXw4v4_X7468rOichbD2PWNkho_IwIsfo'
    print url
    response = urllib2.urlopen(url)
    result = json.load(response)
    try:
        location = result['results'][0]['geometry']['location']
        latitude, longitude = location['lat'], location['lng']
        return latitude, longitude
    except:
        latitude, longitude = None, None
        return latitude, longitude
```

```python
# Getting List of Addresses from 'Line 1 Street Address' and adding 'San Francisco' to them
list_of_addresses = sf_npi_data[u'Line 1 Street Address'].drop_duplicates()
addresses = pd.DataFrame(list_of_addresses)
addresses[u'Line 1 Street Address'] = sf_npi_data[u'Line 1 Street Address'].apply(lambda x: x + ' San Francisco')
```

```python
# Mapping all addresses to coordinates
addresses['coordinates'] = addresses[u'Line 1 Street Address'].apply(lambda x: address_to_coordinates(x))
addresses.to_csv("Addresses.csv")
```
